```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
m = 0
b = 0
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Goal: 
To find the  best  fiiting line through the following data

## Some Formulas
mean squared  error/cost function = (1/n)(sum(y - y_pred)^2);
<br>
y_pred = mx+b
<br>
partial derivative wrt m = (-2/n)*(sum(x(y-(mx+b))))
<br>
partial derivative wrt b = (-2/n)*(sum(y-(mx+b)))
<br>
<br>
m = m - learning rate * partial derivate wrt m
<br>
b = b - learning rate * partial derivate wrt b


```python
def gradient_descent(x,y):
    #initialisation
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.05
    
    for i in range(iterations):
        y_pred = m_curr * x + b_curr
        cost = (1/n)*(sum(val**2 for val in (y-y_pred)))
        md = -(2/n)*(sum(x*(y-y_pred)))
        bd = -(2/n)*(sum(y-y_pred))
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m{}, b{}, cost{}, iterations{}".format(m_curr, b_curr, cost,  i))
        
    plt.scatter(x,y)
    plt.plot(x, m_curr*x + b_curr)
    plt.show()
    
        
```

```python
gradient_descent(x,y)
```
